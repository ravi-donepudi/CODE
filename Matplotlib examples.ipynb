```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Generating samples from a distribution

### A sample of size 10000 drawn from a uniform distribution on [0,100)

```python
sample1 = np.random.uniform(0,100,1000000);
plt.hist(sample1, bins = 100)
plt.show()
```

### Four samples of various sizes drawn from a standard normal distribution

```python
sample2 = np.random.normal(0,1,100)
sample3 = np.random.normal(0,1,1000)
sample4 = np.random.normal(0,1,10000)
sample5 = np.random.normal(0,1,100000)

figs, axs = plt.subplots(2,2,sharey=False)
axs[0][0].hist(sample2, bins=100)
axs[0][0].set_title('100 data points')
axs[0][1].hist(sample3, bins=100)
axs[0][1].set_title('1000 data points')
axs[1][0].hist(sample4, bins=100)
axs[1][0].set_title('10000 data points')
axs[1][1].hist(sample5, bins=100)
axs[1][1].set_title('100000 data points')
for ax in axs.flat:
    ax.label_outer()
figs.suptitle('Samples of various sizes drawn from a Normal distribution of mean 0 and variance 1')

plt.show()
```

## Generating random data, adding noise, fitting a linear regression and  

```python
#Generate 100 values between 0,100 and multiply them by 15 and add 9 and normal noise.
x_values = np.random.uniform(0,100,100).reshape(-1,1)
rand_values = np.random.normal(0,10,100).reshape(-1,1)
y_values = x_values*15+19+rand_values.reshape(-1,1)
#Fit a linear regression to recover the coefficient and regression
model = LinearRegression()
reg=model.fit(X=x_values,y=y_values)
slope,intercept =model.coef_, model.intercept_
plt.scatter(x_values,y_values,color='red',linewidths=1)
plt.plot(x_values, slope*x_values+intercept, color='blue', linewidth=0.5)
plt.show()
#R^2 values
print('The model is y=','%.2f*x + %.2f'% (slope[0][0],intercept[0]))
print('The R^2 value is ', reg.score(x_values,y_values))
```
