# Reversing the first k bits of every 2k size chunk of a string

```python
class Solution:
    def reverseStr(self, s: str, k: int) -> str:
        n=len(s)
        final_string=''
        if n<2*k:
            return reverse(s)        
        else:
            return reversefirsthalf(s[0:2*k],k)+reverseStr(s[2*k:n],k)


    def reverse(self,s):
        new_string=''
        for i in range(len(s)):
            new_string+=s[len(s)-i-1]
        return new_string

    def reversefirsthalf(self, s, k):
        s_1=s[0:k]
        s_2=s[k:2*k]
        return reverse(s_1)+s_2  




```

# ROTATING an nxn matrix in place


```python
#Main idea. Under clockwise rotation index (n-1-j,i) becomes (j,i).
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        n=len(matrix)
        self.transpose(matrix,n)
        self.switcharoo(matrix,n)


    def transpose(self,matrix,n):
        for i in range(n):
            for j in range(i,n):
                matrix[i][j], matrix[j][i]=matrix[j][i], matrix[i][j]

    def switcharoo(self,matrix,n):    
        for i in range(n):
            matrix[i]=[matrix[i][j] for j in range(len(matrix[i])-1,-1,-1)]






matrix=[[1,2,3],[4,5,6],[7,8,9]];
sol=Solution()
sol.rotate(matrix);
print(matrix)

```
