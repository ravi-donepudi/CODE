```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if needle=="":
            return 0
        if haystack=="" or len(needle)>len(haystack):
            return -1
        for i in range(len(haystack)-len(needle)+1):
            if haystack[i]!=needle[0]:
                continue
            if haystack[i:len(needle)+i]==needle:
                return i
        return -1
        
```

```python
class Solution:
    def countAndSay(self, n: int) -> str:
        if n==1:
            print('1','1')
            return '1'
        def reader(word):
            if len(word)==0:
                return ''
            if len(word)==1:
                return '1'+word
            answer = ''
            current, count =word[0], 1
            for c in word[1:]:
                if(c==current):
                    count+=1
                else:
                    answer+=str(count)+str(current)
                    current, count = c, 1
            return answer+str(count)+str(current)
        temp=reader(self.countAndSay(n-1))
        print(n, temp)
        return temp
sol = Solution()
sol.countAndSay(38)
```
